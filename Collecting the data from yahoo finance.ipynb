```python
#importing y finance
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
dataGoogle =  yf.download( 'GOOG','2018-01-01','2019-01-01')
dataFB=  yf.download( 'FB','2018-01-01','2019-01-01')
dataaAAPL =  yf.download( 'AAPL','2018-01-01','2019-01-01')
dataAMZn =  yf.download( 'AMZN','2018-01-01','2019-01-01')
```

```python
#visualising the stock data of Apple
data_apple = dataaAAPL.index
plt.plot(data_apple,dataaAAPL.Close,label = "Close")
plt.plot(data_apple,dataaAAPL.Open,label = "Open")
plt.legend(title = 'APPLE INC')
fig  = plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```

```python
#visualising the stock data of google
data_google = dataGoogle.index
plt.plot(data_google,dataGoogle.Close,label = "Close")
plt.plot(data_google,dataGoogle.Open,label = "Open")
plt.legend(title = 'GOOGLE Inc')
fig  = plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```

```python
#Visualising the stock data of Facebook
data_fb = dataFB.index
plt.plot(data_fb,dataFB.Close,label = "Close")
plt.plot(data_fb,dataFB.Open,label = "Open")
plt.legend(title = 'FACEBOOK Inc')
fig  = plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```

```python
#visualising the stock data of Amazon
data_amazon = dataAMZn.index
plt.plot(data_fb,dataAMZn.Close,label = "Close")
plt.plot(data_fb,dataAMZn.Open,label = "Open")
plt.legend(title = 'AMAZON Inc')
fig  = plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```

```python
# plotting all the companies data
date = dataAMZn.index
plt.plot(date,dataGoogle.High,label = "Google")
plt.plot(date,dataaAAPL.High,label = "Apple")
plt.plot(date,dataFB.High,label = "Facebook")
plt.plot(date,dataAMZn.High,label = "Amazon")
plt.legend(title = "High Price")
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```

```python
#volume plot
date = dataAMZn.index
plt.plot(date,dataGoogle.Volume,label = "Google")
plt.plot(date,dataaAAPL.Volume,label = "Apple")
plt.plot(date,dataFB.Volume,label = "Facebook")
plt.plot(date,dataAMZn.Volume,label = "Amazon")
plt.legend(title = "VOlume")
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```
